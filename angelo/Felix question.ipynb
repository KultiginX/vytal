```python
import pandas as pd
```

```python
df = pd.read_csv(r'container_movements_felix.csv')

df_internal_accounts = pd.read_csv(r'internal_accounts.csv')


df_adm_acc = pd.read_csv(r'container_movements_internal_admin_acc_felix.csv')



#the below mentioned dataset shows the transfers from the same store to itself!!!
df_store_to_store = pd.read_csv(r'container_movements_same_store_movements_inside_store.csv')
```

```python
df_merchants = pd.read_csv(r'Merchant_info.csv')
```

```python
df_MerchantStoreAppUserLinks = pd.read_csv(r'MerchantStoreAppUserLinks_felix.csv')
```

```python
df['OwnerTill'] = pd.to_datetime(df['OwnerTill'])
```

```python
df_checkout = pd.read_csv(r'Total_checkouts_all_time.csv')
```

```python
# first get all the MerchantStoreIds of these stores 
df_MerchantStoreAppUserLinks_merged = df_MerchantStoreAppUserLinks[df_MerchantStoreAppUserLinks.UserId.isin(df['NextOwnerId'].unique())]

# merge with the df_merchants STOREID
df_MerchantStoreAppUserLinks_merged_stores = df_merchants[df_merchants.Id.isin(df_MerchantStoreAppUserLinks_merged.MerchantStoreId.unique())]

df_MerchantStoreAppUserLinks_merged_stores = df_MerchantStoreAppUserLinks_merged_stores[df_MerchantStoreAppUserLinks_merged_stores.Id != '603a043e-c3e7-4550-99c1-4c21bc66e93e']

df_MerchantStoreAppUserLinks_merged_stores



```

```python
df_MerchantStoreAppUserLinks_merged = df_MerchantStoreAppUserLinks[df_MerchantStoreAppUserLinks.UserId.isin(df['NextOwnerId'].unique())]


df_MerchantStoreAppUserLinks_merged_stores = pd.merge(left = df_merchants, 
                                                      right = df_MerchantStoreAppUserLinks_merged, 
                                                      left_on = 'Id', 
                                                      right_on = 'MerchantStoreId', 
                                                      how = 'inner')


df_MerchantStoreAppUserLinks_merged_stores = df_MerchantStoreAppUserLinks_merged_stores[\
                                                        df_MerchantStoreAppUserLinks_merged_stores.Id \
                                                        != '603a043e-c3e7-4550-99c1-4c21bc66e93e']
```

```python
# now add those other stores that are in the df_MerchantStoreAppUserLinks; merge their USERIDs on their merchant ids -- in the end merge on df_merchants (to get names)
# simply merge the MerchantStoreId with the MerchantId in the df_merchants 
df['acc_to'] = 'nan'

# user and adm acc
df.loc[df.NextOwnerId.isin(df_adm_acc.To.unique()), 'acc_to'] = 'adm_acc'
df.loc[df.NextOwnerId.isin(df_checkout.UserId.unique()), 'acc_to'] = 'user_acc'

df.loc[df.NextOwnerId.isin(df_MerchantStoreAppUserLinks_merged_stores.UserId.unique()), 'acc_to'] = 'diff_store'
df.loc[df.NextOwnerId.isin(df_internal_accounts.Id.unique()), 'acc_to'] = 'int_acc'

df.loc[df.NextOwnerId == '66a00c85-ba23-4c84-b41e-95e04dd9d12e', 'acc_to'] = 'same_store'
```

```python
df['OwnerTill'] = df['OwnerTill'].astype(str) 
df['OwnerSince'] = df['OwnerSince'].astype(str)
```
